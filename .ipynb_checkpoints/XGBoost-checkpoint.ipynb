## One split for t = 1 (first tree)

```python
rng = np.random.default_rng()
x = np.linspace(0, 10, 100)
x = []
for i in range(0,100):
    n = random.uniform(1,10)
    x.append(n)
x = np.array(x)


y = np.where(x < 5, x,5) + rng.normal(0, 0.3, size=x.shape)
#x = x.reshape(-1,1)

df = pd.DataFrame({"X":x,"y":y})

plt.scatter(df["X"],df["y"], label='y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

```python
f_0 = np.mean(y)

g = []
h = []

#Compute gradient and hessian of each sample
df["g"] = 2*(df["y"]-f_0)/len(df["y"])
df["h"] = -2/len(df["y"])

df = df.sort_values(by=['X'])
df.reset_index(inplace = True, drop = True)

opt_split = 0
x_opt = 0

#Keep track of the loss
l = []

#Test the different threshold
for i in range(len(df)):
    t = df["X"][i]
    df_L =  df[df["X"]<t]
    df_R =  df[df["X"]>=t]
    
    G = np.sum(df["g"])
    H = np.sum(df["h"])
    
    G_L = np.sum(df_L["g"])
    G_R = np.sum(df_R["g"])
    
    H_L = np.sum(df_L["h"])
    H_R = np.sum(df_R["h"])
    
    if (H==0) or (H_R==0) or (H_L ==0):
        continue
        
    split = 1/2*((G_L*G_L)/H_L + (G_R*G_R)/H_R - (G*G)/H)
    
    l.append(split)
    
    opt_split = min(opt_split, split)
    
    if split == opt_split:
        x_opt = df["X"][i]
    
```

```python
plt.plot(df["X"][1:],l)
plt.title("Split impact on loss for different threshold values")
plt.xlabel("Threshold value")
plt.ylabel("Split impact")
```

```python
x_opt
```

This is the optimal value on which to split

```python
#Compute the new leaf weights
df_left = df[df["X"]<x_opt]
df_right = df[df["X"]>x_opt]


G_Left = np.sum(df_left["g"])
G_Right = np.sum(df_right["g"])
    
H_Left = np.sum(df_left["h"])
H_Right = np.sum(df_right["h"])

w_left = -G_Left/H_Left
w_right = -G_Right/H_Right
```

```python
w_left
```

For x<x_opt this is the correction done to the previous prediction

```python
w_right
```

For x>x_opt this is the correction done to the previous prediction

```python
f_1 = []

for i in df["X"]: 
    if i<x_opt:
        f_1.append(f_0 + w_left)
    else :
        f_1.append(f_0 + w_right)

plt.figure()
plt.scatter(df["X"],df["y"])
plt.plot(df["X"],f_1, color = "red")
```

## XGBoost with hist - one client

```python
rng = np.random.default_rng()

x = []
for i in range(0,100):
    n = random.uniform(1,10)
    x.append(n)
x = np.array(x)



y = np.where(x < 5, x,5) + rng.normal(0, 0.3, size=x.shape)

df = pd.DataFrame({"X":x,"y":y})

df = df.sort_values(by=['X'])
df.reset_index(inplace = True, drop = True)


f_0 = np.mean(df["y"])
f_m = f_0

trees = []
for t in range(1,10):
    #Compute gradient and hessian of each sample
    df["g"] = 2*(df["y"]-f_m)/len(df["y"])
    df["h"] = -2/len(df["y"])

    df_bin = df.copy()

    df_bin.drop("y",axis = 1, inplace = True)
    #How many bins you want
    b = 20

    df_bin["bin"] = pd.qcut(df_bin["X"],b)

    df_bin = df_bin.groupby('bin', as_index=False)[['g','h']].sum()
    df_bin["X"] = df_bin["bin"].apply(lambda x :x.mid)

    df_bin.drop("bin",axis = 1, inplace = True)

    df_bin = df_bin[["X","g","h"]]

    #Build the tree
    tree = split(df_bin, 0, 2)
    trees.append(tree)

    #Predict the residuals 
    preds = df.apply(predict, axis='columns', rules=tree.copy())


    #New model predictions
    f_m = f_m + preds

    plt.figure()
    plt.scatter(df["X"],df["y"])
    plt.plot(df["X"],f_m, color = "red")
```

## XGBoost with hist - several clients 

```python
clients = []

for i in range(0,20):
    rng = np.random.default_rng()
    x = np.linspace(0, 10, 50)
    x = []
    for i in range(0,100):
        n = random.uniform(1,10)
        x.append(n)
    x = np.array(x)
    
    y = np.where(x < 5, x,5) + rng.normal(0, 0.3, size=x.shape)

    df = pd.DataFrame({"X":x,"y":y})
    
    df = df.sort_values(by=['X'])
    df.reset_index(inplace = True, drop = True)
    
    clients.append(df)
    
print("Done")
f_0 = 0
f_m = [f_0]*len(clients)

plt.figure()
plt.scatter(clients[0]["X"],clients[0]["y"])
plt.plot(clients[0]["X"],[f_m[0]]*len(clients[0]), color = "red")

trees = []
for t in range(1,10):
    binned_client = []

    for i in range(len(clients)):
        #Compute gradient and hessian of each sample
        clients[i]["g"] = 2*(clients[i]["y"]-f_m[i])/len(clients[i]["y"])
        clients[i]["h"] = -2/len(clients[i]["y"])

        
        df_bin = clients[i].copy()

        df_bin.drop("y",axis = 1, inplace = True)
        #How many bins you want
        b = 20

        df_bin["bin"] = pd.qcut(df_bin["X"],b)

        df_bin = df_bin.groupby('bin', as_index=False)[['g','h']].mean()
        df_bin["X"] = df_bin["bin"].apply(lambda x :x.mid)

        df_bin.drop("bin",axis = 1, inplace = True)

        df_bin = df_bin[["X","g","h"]]

        binned_client.append(df_bin)


    aggregator = binned_client[0]
    for i in range(1, len(clients)):
        aggregator = aggregator.append(binned_client[i])

    agg = merge_hist(aggregator, 20)

    #Build the tree
    tree = split(agg, 0, 2)

    trees.append(tree)
    
    for i in range(len(clients)):
        preds = clients[i].apply(predict, axis='columns', rules=tree.copy())
        f_m[i] = f_m[i]+preds



    plt.figure()
    plt.scatter(clients[0]["X"],clients[0]["y"])
    plt.plot(clients[0]["X"],f_m[0], color = "red")

    
```

## Histogram merging function 

```python
def merge_hist(df,k):
    m = len(df)
    if k>m:
        print("already satisfied")
        return df
    else:
        while k<m:
            df = df.sort_values(by=['X'])
            df.reset_index(inplace = True, drop = True)
            a = list(df["X"])
            b = [j-i for i, j in zip(a[:-1], a[1:])]

            mini = b.index(min(b))

            new = [(a[mini]+a[mini+1])/2,df["g"][mini]+df["g"][mini+1],df["h"][mini]+df["h"][mini+1]]

            df.loc[len(df)] = new
            df.drop([mini,mini+1], inplace = True)
            m = len(df)
        return df
```

## Tree functions

```python
def find_best_rule(X_train):
    
    #X_train with X, g and h for all samples
    
    best_feature, best_threshold, opt_split = None, None, np.inf
    df = X_train.copy()
    for feature in list(X_train.columns)[:-2]:
        thresholds = X_train[feature].unique().tolist()
        thresholds.sort()
        thresholds = thresholds[1:]
        for t in thresholds:
            df_L =  df[df["X"]<t]
            df_R =  df[df["X"]>=t]
    
            G = np.sum(df["g"])
            H = np.sum(df["h"])
    
            G_L = np.sum(df_L["g"])
            G_R = np.sum(df_R["g"])
    
            H_L = np.sum(df_L["h"])
            H_R = np.sum(df_R["h"])
            
              
            if (H==0) or (H_R==0) or (H_L ==0):
                continue
        
            split = 1/2*((G_L*G_L)/H_L + (G_R*G_R)/H_R - (G*G)/H)
            
            opt_split = min(opt_split, split)
            
            if split == opt_split:
                x_opt = t
                best_feature = feature
    
    return {'feature': best_feature, 'threshold': x_opt}



def split(X_train, depth, max_depth):
    if depth == max_depth or len(X_train) < 2:
        return {'prediction': -np.sum(X_train["g"])/np.sum(X_train["h"])}
    
    
    rule = find_best_rule(X_train)
    left_ix = X_train[rule['feature']] < rule['threshold']
    rule['left'] = split(X_train[left_ix], depth + 1, max_depth)
    rule['right'] = split(X_train[~left_ix], depth + 1, max_depth)
    return rule

def predict(sample, rules):
    prediction = None
    while prediction is None:
        feature, threshold = rules['feature'], rules['threshold']
        if sample[feature] < threshold:
            rules = rules['left']
        else:
            rules = rules['right']
        prediction = rules.get('prediction', None)
    return prediction


def evaluate(X, y):
    preds = X.apply(predict, axis='columns', rules=rules.copy())
    return r2_score(preds, y)


def rss(y_left, y_right):
    def squared_residual_sum(y):
        return np.sum((y - np.mean(y)) ** 2)
    
    return squared_residual_sum(y_left) + squared_residual_sum(y_right) 
```
